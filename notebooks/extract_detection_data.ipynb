```python
import os
```

```python
def parse_yolo_output(txt_file_path):
    detections = []
    with open(txt_file_path, 'r') as f:
        for line in f:
            label, x_center, y_center, width, height, confidence = line.strip().split()
            detections.append({
                'label': label,
                'x_center': float(x_center),
                'y_center': float(y_center),
                'width': float(width),
                'height': float(height),
                'confidence': float(confidence)
            })
    return detections
```

```python
detection_files = r'C:\Users\ftesfaye\Desktop\KIFIYA\kifiya_week_7\Data-Warehouse-for-Ethiopian-Medical-Business-Data\yolov5\runs\detect\exp\labels'
```

```python
for txt_file in os.listdir(detection_files):
    txt_file_path = os.path.join(detection_files, txt_file)
    detections = parse_yolo_output(txt_file_path)
    print(f"Detections for {txt_file}:")
    print(detections)
```

```python
import psycopg2
import pandas as pd
```

```python
conn = psycopg2.connect(
    database="medical", 
    user="postgres", 
    password="admin", 
    host="localhost", 
    port="5432"
)

# Create a cursor object
cur = conn.cursor()
```

```python
cur.execute('''
    CREATE TABLE IF NOT EXISTS object_detection_results (
        detection_id SERIAL PRIMARY KEY,
        image_name VARCHAR(255),
        label VARCHAR(100),
        x_center FLOAT,
        y_center FLOAT,
        width FLOAT,
        height FLOAT,
        confidence FLOAT
    )
''')
conn.commit()
```

```python
def insert_detection_data(image_name, detections):
    for detection in detections:
        cur.execute('''
            INSERT INTO object_detection_results (
                image_name, label, x_center, y_center, width, height, confidence
            ) VALUES (%s, %s, %s, %s, %s, %s, %s)
        ''', (
            image_name, detection['label'], detection['x_center'], detection['y_center'],
            detection['width'], detection['height'], detection['confidence']
        ))
    conn.commit()

# Path to the folder containing YOLO detection text files
detection_files_path = r'C:\Users\ftesfaye\Desktop\KIFIYA\kifiya_week_7\Data-Warehouse-for-Ethiopian-Medical-Business-Data\yolov5\runs\detect\exp\labels'

# Loop through the text files, parse them, and store results in the database
for txt_file in os.listdir(detection_files_path):
    txt_file_path = os.path.join(detection_files_path, txt_file)
    image_name = os.path.splitext(txt_file)[0]  # Get image name without extension
    detections = parse_yolo_output(txt_file_path)
    insert_detection_data(image_name, detections)
    print(f"Inserted detections for {image_name}")

```

```python
# Close the connection
cur.close()
conn.close()
```
